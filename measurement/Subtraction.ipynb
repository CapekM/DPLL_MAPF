```python
# filename = "grid_50_obs_30.csv"
# filename = "grid_32.csv"
# filename = "grid_16.csv"
# filename = "maze_32.csv"
filename = "warehouse.csv"
# filename = "grid_20_obs_30.csv"

data = pd.read_csv(filename, index_col=0)
data = data.set_index("name")
data
```

```python
# Sum of percentage improvement. Minus means deterioration.
df = data.copy()
for n in ["3", "5", "10", "EXP 1.4", "EXP 1.6", "EXP"]:
    df[n] = df["SMT-CBS"] - df[n]
df.loc["Total"] = df.sum()
df
```

```python
df.to_csv(f"sub/{filename}")
```

```python
def load_output(filename, n):
    p = list(range(10))
    p.insert(0, "name")
    data = pd.read_csv(filename, names=p)
    data[n] = data.mean(numeric_only=True, axis=1)
    data = data.drop(list(range(10)), axis=1)
    return data
```

```python
data2 = load_output("../code/output/dpll_mapf_exp_14.csv", "EXP 1.4")
data3 = load_output("../code/output/dpll_mapf_exp_16.csv", "EXP 1.6")
data2
```

```python
data = pd.merge(data,data2,on=["name"])
data = pd.merge(data,data3,on=["name"])
data
```
