```python
import numpy as np
import pandas as pd
```

```python
# name = "maze"
# name = "grid_32"
name = "warehouse"
```

```python
data = pd.read_csv("average/"+name+".csv", index_col=0)
del data["SMT-CBS"]
data = data.set_index("name")
data
```

```python
data_c = pd.read_csv("choke_experiment/"+name+".csv", index_col=0)
data_c = data_c.set_index("name")
data_c
```

```python
df = data_c.copy()
for n in ["3", "5", "10", "EXP 1.4", "EXP 1.6", "EXP"]:
    df[n] = data[n] - df[n]
df.loc["Total"] = df.sum()
df
```

```python
df.to_csv(f"choke_experiment/{name}_sub.csv")
```
